```python
import tkinter as tk

class ElectricityBill:
    def __init__(self, customer_name, customer_address, meter_number, prev_reading, curr_reading):
        self.customer_name = customer_name
        self.customer_address = customer_address
        self.meter_number = meter_number
        self.prev_reading = prev_reading
        self.curr_reading = curr_reading

    def calculate_consumption(self):
        return self.curr_reading - self.prev_reading

    def calculate_bill(self, rate):
        consumption = self.calculate_consumption()
        return consumption * rate

    def generate_bill(self, rate):
        consumption = self.calculate_consumption()
        bill_amount = self.calculate_bill(rate)
        bill_details = f"Customer Name: {self.customer_name}\nAddress: {self.customer_address}\nMeter Number: {self.meter_number}\nPrevious Reading: {self.prev_reading}\nCurrent Reading: {self.curr_reading}\nConsumption: {consumption}\nBill Amount: {bill_amount}"
        return bill_details


def generate_bill():
    customer_name = name_entry.get()
    customer_address = address_entry.get()
    meter_number = meter_entry.get()
    prev_reading = float(prev_reading_entry.get())
    curr_reading = float(curr_reading_entry.get())
    tariff_rate = float(tariff_entry.get())

    bill = ElectricityBill(customer_name, customer_address, meter_number, prev_reading, curr_reading)
    generated_bill = bill.generate_bill(tariff_rate)
    bill_text.delete(1.0, tk.END)
    bill_text.insert(tk.END, generated_bill)


# Create the main window
root = tk.Tk()
root.title("Electricity Billing System")

# Create and place labels and entry fields
tk.Label(root, text="Customer Name:").grid(row=0, column=0, sticky="w")
name_entry = tk.Entry(root)
name_entry.grid(row=0, column=1)

tk.Label(root, text="Address:").grid(row=1, column=0, sticky="w")
address_entry = tk.Entry(root)
address_entry.grid(row=1, column=1)

tk.Label(root, text="Meter Number:").grid(row=2, column=0, sticky="w")
meter_entry = tk.Entry(root)
meter_entry.grid(row=2, column=1)

tk.Label(root, text="Previous Reading:").grid(row=3, column=0, sticky="w")
prev_reading_entry = tk.Entry(root)
prev_reading_entry.grid(row=3, column=1)

tk.Label(root, text="Current Reading:").grid(row=4, column=0, sticky="w")
curr_reading_entry = tk.Entry(root)
curr_reading_entry.grid(row=4, column=1)

tk.Label(root, text="Tariff Rate:").grid(row=5, column=0, sticky="w")
tariff_entry = tk.Entry(root)
tariff_entry.grid(row=5, column=1)

# Create a button to generate the bill
generate_button = tk.Button(root, text="Generate Bill", command=generate_bill)
generate_button.grid(row=6, column=0, columnspan=2)

# Create a text widget to display the generated bill
bill_text = tk.Text(root, height=10, width=50)
bill_text.grid(row=7, column=0, columnspan=2)

root.mainloop()
```
